```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("GoI_yield_cpi.xlsx")

df["Inflation_pct"] = df["CPI"].pct_change() * 100

df["Real_Rate"] = df["Yield"] - df["Inflation_pct"]

plt.figure(figsize=(8,5))
plt.scatter(df["Inflation_pct"], df["Real_Rate"], color="purple")
plt.axhline(0, color="gray", linestyle="--")
plt.xlabel("Annual CPI Inflation (%)")
plt.ylabel("Real Effective Interest Rate (%)")
plt.title("Inflation vs Real Effective Interest Rate on Public Debt (2012â€“2023)")
plt.grid(True)
plt.show()
```
